# 🧠 Phase 3-7: Model Development & Training
## Building and Training Deep Learning Models for Leaf Disease Classification

```python
import sys
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
import numpy as np

sys.path.insert(0, str(Path.cwd().parent))

from config import DATA_DIR, BATCH_SIZE, INITIAL_LR, EPOCHS, CLASS_NAMES
from src.data.dataset import LeafDataset
from src.data.splits import create_dataset_splits
from src.data.augmentation import create_tf_dataset
from src.models.factory import get_model
from src.training.trainer import ModelTrainer
from src.utils.visualization import Visualizer

viz = Visualizer()
```

```python
# Load and prepare data
dataset = LeafDataset(DATA_DIR)
dataset.investigate()

import config
config.CLASS_NAMES = dataset.class_names
config.NUM_CLASSES = dataset.num_classes

clean_paths = dataset.get_clean_file_paths()
train_p, val_p, test_p, train_l, val_l, test_l = create_dataset_splits(clean_paths)
```

```python
# Create data pipelines
train_ds = create_tf_dataset(train_p, train_l, BATCH_SIZE, training=True)
val_ds = create_tf_dataset(val_p, val_l, BATCH_SIZE, training=False)
test_ds = create_tf_dataset(test_p, test_l, BATCH_SIZE, training=False)
```

```python
# Build and train Custom CNN
model = get_model('CustomCNN', learning_rate=INITIAL_LR)
model.summary()
```

```python
# Train
trainer = ModelTrainer(model, 'CustomCNN')
history = trainer.train(train_ds, val_ds, epochs=EPOCHS)
```

```python
# Evaluate
results = trainer.evaluate(test_ds)
print(f"Best val_accuracy: {max(history.history['val_accuracy']):.4f}")
```

```python
# Training curves
viz.plot_training_history(history, 'CustomCNN')
```
